###### Exercise 1.04: Continuing Verification of Data Integrity

```python
id_counts.head()
```

```python
dupe_mask = id_counts == 2
dupe_mask[:5]
```

```python
id_counts.index[:5]
```

```python
dupe_ids = list(id_counts.index[dupe_mask])
print(len(dupe_ids))
dupe_ids[:5]
```

```python
df.loc[df['ID'].isin(dupe_ids),:]
```

```python
df_zero_mask = df == 0
df_zero_mask
```

```python
feature_zero_mask = df_zero_mask.iloc[:,1:].all(axis=1)
sum(feature_zero_mask)
```

```python
df_clean_1 = df.loc[~feature_zero_mask,:].copy()
len(df_clean_1)
```

```python
df_clean_1.shape, df_clean_1['ID'].nunique()
```

```python
# df_clean_1.to_csv('Datasets/df_clean_1.csv', index=False)
```

###### Exercise 1.05: Exploring and Cleaning the Data 

```python
import pandas as pd
```

```python
df_clean_1['PAY_1'].value_counts()
```

###### Deep Dive: Categorical Features

###### Exercise 1.07: Implementing OHE for a Categorical Feature

```python
import pandas as pd
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 400
df_clean_2 = pd.read_csv('Datasets/df_clean_2_01.csv')
df_clean_2.head()
```

```python
df_clean_2['EDUCATION_CAT'] = None
df_clean_2.head()
```

```python
df_clean_2[['EDUCATION', 'EDUCATION_CAT']].head(10)
```

```python
cat_mapping = {1: 'graduate school', 2: 'university', 3: 'high school', 4: 'others'}
df_clean_2['EDUCATION_CAT'] = df_clean_2['EDUCATION'].map(cat_mapping)
df_clean_2[['EDUCATION', 'EDUCATION_CAT']].head(10)
```
